### Problem 1

```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import random as rnd
```

```python
def problem_1(n):
    y = np.empty(1)
    x = rnd.normal(size=(n,n))
    avgs = np.mean(x, axis=0)
    y = np.var(avgs)
    return y
```

```python
def problem_1_1():
    results = np.empty(10) #initialise array
    for i in range(100, 1000, 100):
        q = int(i/100)
        results[q] = problem_1(i)
    return results
```

```python
plot_array = problem_1_1()

plt.plot(plot_array)
plt.show()
```

### Problem 2

```python
def problem_2():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y_sin = np.sin(x)
    y_cos = np.cos(x)
    y_arctan = np.arctan(x)
    
    plt.plot(x, y_sin)
    plt.plot(x, y_cos)
    plt.plot(x, y_arctan)
    
    plt.show()
    return

problem_2()
```

### Problem 3

```python
def problem_3():
    x_left = np.linspace(-2, 0.95, 300)
    x_right = np.linspace(1.05, 6, 500)
    y_left = 1/(x_left -1)
    y_right = 1/(x_right -1)
    
    plt.plot(x_left, y_left, 'm--', linewidth = 4, )
    plt.plot(x_right, y_right, 'm--', linewidth = 4,)
    
    plt.xlim(-2, 6)
    plt.ylim(-6, 6)
    
    plt.show()

problem_3()
```

### Problem 4

```python
def problem_4():
    x = np.linspace(0, 2*np.pi, 100)
    y_1 = np.sin(x)
    y_2 = np.sin(2*x)
    y_3 = 2*np.sin(x)
    y_4 = 2*np.sin(2*x)
    
    ax1 = plt.subplot(221)
    ax1.plot(x, y_1, 'g-', linewidth=4)
    ax1.set_title('sin(x)')
    ax1.set_xlim(0, 2*np.pi)
    ax1.set_ylim(-2, 2)
    
    ax2 = plt.subplot(222)
    ax2.plot(x, y_2, 'r--', linewidth=4)
    ax2.set_title('sin(2x)')
    ax2.set_xlim(0, 2*np.pi)
    ax2.set_ylim(-2, 2)
    
    ax3 = plt.subplot(223)
    ax3.plot(x, y_3, 'b--', linewidth=4)
    ax3.set_title('2sin(x)')
    ax3.set_xlim(0, 2*np.pi)
    ax3.set_ylim(-2, 2)
    
    ax4 = plt.subplot(224)
    ax4.plot(x, y_4, 'm:', linewidth=4)
    ax4.set_title('2sin(2x)')
    ax4.set_xlim(0, 2*np.pi)
    ax4.set_ylim(-2, 2)
    
    plt.suptitle('Trig graphs')
    
    plt.show()
    return

problem_4()
```

### Problem 5

```python
FARS = np.load('FARS.npy') #load the data

def problem_5():
    
    plt.figure(figsize=(10, 3))
    
    ax1 = plt.subplot(121)
    ax1.plot(FARS[:,1], FARS[:,2], 'k*')
    ax1.set_xlabel('Longitude')
    ax1.set_ylabel('Latitude')
    ax1.set_aspect('equal')
    
    ax2 = plt.subplot(122)
    ax2.hist(FARS[:,0], bins=23, range=[0, 24])
    ax2.set_xlabel('Time (24 hours)')
    ax2.set_xlim(0, 24)
    
    plt.show()
    return

problem_5()

```

### Problem 6

```python
def problem_6():
    
    plt.figure(figsize=(10, 3))
    
    #Heat map
    x = np.linspace(-2*np.pi, 2*np.pi, 1000) #set up domain of x
    y = np.linspace(-2*np.pi, 2*np.pi, 1000) #set up domain of y
    X, Y = np.meshgrid(x, y) #extract values of domain
    Z = (np.sin(X)*np.sin(Y))/(X*Y) #compute output of function
    
    plt.subplot(121)
    plt.pcolormesh(X, Y, Z, cmap='Reds')
    plt.colorbar()
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(-2*np.pi, 2*np.pi)
    
    #Contour map
    plt.subplot(122)
    plt.contour(X, Y, Z, 10, cmap='hot')
    plt.colorbar()
    
    return

problem_6()
```
