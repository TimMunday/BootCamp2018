### Arbuthnot

```python
from pydataset import data
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Accent')
births = data('Arbuthnot')

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.plot(births.Year, births.Ratio, color=palette(0), linewidth = 2.0)
plt.xlabel('Year')
plt.ylabel('Ratio of male to female births')

plt.subplot(2, 2, 2)
plt.plot(births.Year, births.Total, color=palette(1), linewidth = 2.0)
plt.xlabel('Year')
plt.ylabel('Total no of births')

plt.subplot(2, 2, 3)
plt.scatter(births.Total, births.Ratio, color=palette(2), linewidth = 2.0)
plt.xlabel('Total no of births')
plt.ylabel('Ratio of male to female births')

plt.show()
```
