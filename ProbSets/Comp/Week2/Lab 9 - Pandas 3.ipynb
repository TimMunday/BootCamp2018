```python
import numpy as np
import pandas as pd
from numpy import random as rand
from matplotlib import pyplot as plt
from pydataset import data


plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Accent')
```

### Iris

```python
iris = data('iris')

#Plot characteristics by species

spec = iris.groupby('Species')
spec.mean().plot(kind='barh')
plt.show()
```

Virginica is the easiest to identify, it has the largest petals by both length and width. Although Setosa may also be easy to identify based on the ratio of width:length.

### Poisons

```python
poisons = data('poisons')

poison = poisons.groupby('poison')
poison.mean().plot(kind='barh')
plt.show()
```

Clearly poison three is the most deadly.

```python
poison_treat = poisons.groupby('treat')
poison_treat['time'].mean().plot(kind='barh')
plt.xlabel('time')
plt.show()
```

Treatment B is the most effective

```python
poison_subs=poisons.groupby(['poison', 'treat'])
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab20b')

poison_subs.mean().unstack().plot(kind='barh')

plt.show()
```

### Diamonds

```python
diamonds=data('diamonds')

cut = diamonds.groupby('cut')

diamonds.boxplot(['price'], by='cut')
plt.show()
```

Price increases with the quality of the cut

```python
diamonds['volume'] = diamonds['x'] * diamonds['y'] * diamonds['z']
diamonds[diamonds.color == 'H'].groupby('cut')['volume'].mean().plot(kind='bar')
plt.title('Diamond volumes')
plt.show()
```

Bigger diamonds have higher prices, and lower cut quality is positively correlated with volume. This explains why the average price of fair cuts may be higher than those with an ideal cut.
