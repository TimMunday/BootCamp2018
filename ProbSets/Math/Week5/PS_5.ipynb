```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
dom = np.linspace(0, 50, 10000)
c_1 = (2*dom + 4)/3
c_2 = (dom-1)/6
c_3 = 6 - dom

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 10))
palette = plt.get_cmap('Dark2')

plt.plot(dom, c_1, color=palette(1))
plt.plot(dom, c_2, color=palette(0))
plt.plot(dom, c_3, color=palette(4))
plt.axvline(x=0, color=palette(2))
plt.plot(dom, np.zeros(10000), color=palette(3))
plt.fill_between(dom, c_3, c_1, color='grey', where =  c_3 >= c_1)
plt.xlim(-1e-2, 6)
plt.ylim(-1e-2, 6)


#obj = (5x - c)/4
obj_1 = (5*dom - 1) / 4
obj_2 = (5*dom - 5) / 4
obj_3 = (5*dom + 3) / 4

plt.plot(dom, obj_1, 'r--')
plt.plot(dom, obj_2, 'r--')
plt.plot(dom, obj_3, 'r--')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

We can see that the maximum is at the point at which our two binding constraints meet (the objective function increases out and to the bottom right). Solving these two equations gives x = 2.8 and y = 3.2

### Ex 8.2

```python
def intersection(a1, b1, c1, a2, b2, c2):
    """a1x + b1y = c1
       a2x + b2y = c2
    """
    determinant = a1 * b2 - a2 * b1
    if (determinant == 0):
        print("Lines are parallel")

    else:
        x = (c1 * b2 - c2 * b1)/determinant
        y = (a1 * c2 - a2 * c1)/determinant
    return x, y
```

```python
dom = np.linspace(0, 50, 10000)
d_1 = (15-dom)/3 #less than
d_2 = (18-2*dom)/3 #less than
d_3 = dom - 4 #greater than

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 10))
palette = plt.get_cmap('Dark2')

plt.plot(dom, d_1, color=palette(1))
plt.plot(dom, d_2, color=palette(0))
plt.plot(dom, d_3, color=palette(4))
plt.axvline(x=0, color=palette(2))
plt.plot(dom, np.zeros(10000), color=palette(3))
plt.fill_between(dom, d_3, d_1, color='grey', where = d_2 >= d_1, interpolate = True)
plt.fill_between(dom, d_3, d_2, color='grey', where = d_2 <= d_1, interpolate = True)# & d_3 <= d_2)
plt.xlim(-1e-2, 6)
plt.ylim(-1e-2, 8)


#obj = (c-3x)
obj_11 = 1-3*dom
obj_22 = 5-3*dom
obj_33 = 10-3*dom
obj_44 = 15 -3*dom

plt.plot(dom, obj_11, 'r--')
plt.plot(dom, obj_22, 'r--')
plt.plot(dom, obj_33, 'r--')
plt.plot(dom, obj_44, 'r--')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Clearly the optimiser is the vertex 6, 2
