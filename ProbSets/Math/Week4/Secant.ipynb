# Secant

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def secant(x_0, x_1, tol, fdash):
    
    iters = 0
    diff = 5
    x_k = x_1
    x_km1 = x_0
    save = np.zeros(100)
    
    while diff > np.abs(x_k)*tol and iters < 100:
        x_kp1 = x_k - fdash(x_k)*(x_k - x_km1)/(fdash(x_k) - fdash(x_km1))
        save[iters] = x_kp1
        diff = np.abs(x_kp1 - x_k)
        x_km1 = x_k
        x_k = x_kp1
        iters += 1
    
    return x_kp1, iters, save
```

```python
#Check to see if works
def f(x):
    output = 4*x**3 - 42*x**2 +120*x  - 70
    return output
```

```python
#Check
sol, iterations, toplot = secant(0, 0.1, 0.001, f)
print(sol)
print(iterations)
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 10))
dom = np.linspace(-0.2, 1, 1000)
g = dom**4 -14*dom**3 + 60*dom**2 - 70*dom
plt.plot(dom, g)
plt.xlabel('x')
for i in range(iterations + 1):
    plt.plot(toplot[i], -25, marker='o', markersize=8)
plt.show()
```
